1. Базовая модель

Оболочкам Keras требуется функция в качестве аргумента. Эта функция, которую мы должны определить, отвечает за создание модели нейронной сети для оценки.

Ниже мы определяем функцию для создания базовой модели для оценки. Это простая модель, которая имеет один полностью связанный скрытый слой с тем же количеством нейронов, что и входные атрибуты (13). В сети используются полезные практики, такие как функция активации выпрямителя для скрытого слоя. Для выходного слоя функция активации не используется, поскольку это проблема регрессии, и мы заинтересованы в прогнозировании числовых значений напрямую без преобразования.

Используется эффективный алгоритм оптимизации ADAM и оптимизируется функция потери среднего квадрата ошибки. Это будет та же метрика, которую мы будем использовать для оценки производительности модели. Это желательный показатель, потому что получение квадратного корня дает нам значение ошибки, которое мы можем непосредственно понять в контексте проблемы

```python
# define base model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(13, input_dim=13, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

Объект оболочки Keras для использования в scikit-learn в качестве регрессионной оценки называется KerasRegressor. Мы создаем экземпляр и передаем ему как имя функции для создания модели нейронной сети, так и некоторые параметры для передачи функции fit () модели позже, например, количество эпох и размер пакета. Оба из них установлены на разумные значения по умолчанию.

Мы также инициализируем генератор случайных чисел с постоянным случайным начальным числом, процесс, который мы будем повторять для каждой модели, оцененной в этом руководстве. Это попытка обеспечить последовательное сравнение моделей.

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# evaluate model with standardized dataset
estimator = KerasRegressor(build_fn = baseline_model, epochs=100, batch_size=5, verbose=0)
```

Последний шаг - оценить эту базовую модель. Мы будем использовать 10-кратную перекрестную проверку для оценки модели.

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

ыполнение этого кода дает нам оценку производительности модели по проблеме для невидимых данных. В результате сообщается о среднем квадрате ошибки, включая среднее и стандартное отклонение (среднее отклонение) по всем 10-кратным оценкам перекрестной проверки.

2. Стандартизированный набор

Важной проблемой, связанной с набором данных, является то, что все входные атрибуты различаются по своим масштабам, поскольку они измеряют различные количества.

Практически всегда полезно подготовить данные перед их моделированием с использованием модели нейронной сети.

Продолжая вышеприведенную базовую модель, мы можем переоценить ту же модель, используя стандартизированную версию входного набора данных.

Мы можем использовать Scikit-Learn для выполнения стандартизации во время процесса оценки модели, в пределах каждого раза перекрестной проверки. Это гарантирует отсутствие утечки данных из каждого перекрестного валидационного набора тестов в обучающие данные.

Приведенный ниже код создает конвейер scikit-learn, который сначала стандартизирует набор данных, а затем создает и оценивает базовую модель нейронной сети.

```python
# evaluate model with standardized dataset
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=baseline_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, shuffle = True, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

Выполнение примера обеспечивает улучшенную производительность по сравнению с базовой моделью без стандартизированных данных, исключая ошибку.

Еще одним расширением этого раздела будет аналогичное применение масштабирования к выходной переменной, например, его нормализация в диапазоне 0-1, и использование сигмоидальной или аналогичной функции активации на выходном слое для сужения выходных прогнозов в том же диапазоне.
