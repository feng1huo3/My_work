3.2. Отношения функций

```python
df_corr = df.corr(method="spearman")
```

В зависимости от набора данных и типа объектов (например, порядковые или непрерывные объекты) можно использовать spearman-метод вместо pearson-метода для вычисления корреляции. В то время как корреляция Пирсона оценивает линейную связь между двумя непрерывными переменными, корреляция Спирмена оценивает монотонную связь на основе ранжированных значений для каждой функции. И чтобы помочь с интерпретацией этой корреляционной матрицы, давайте воспользуемся Seaborn.heatmap()для ее визуализации.

```python
# Create labels for the correlation matrix
labels = np.where(np.abs(df_corr)>0.75, "S",
                  np.where(np.abs(df_corr)>0.5, "M",
                           np.where(np.abs(df_corr)>0.25, "W", "")))

# Plot correlation matrix
plt.figure(figsize=(15, 15))
sns.heatmap(df_corr, mask=np.eye(len(df_corr)), square=True,
            center=0, annot=labels, fmt='', linewidths=.5,
            cmap="vlag", cbar_kws={"shrink": 0.8});
```

Большой разброс значений потребует масштабирование признаков.

Обнаружены выбросы, которые необходимо изучить подробнее:

Длинный хвост распределения влево и вправо может стать проблемой для алгоритмов МО.
Некоторые признаки стремятся к колоколообразному распределнию для целей МО.

При этом распределение значений признака поверхностная плотность вызывает вопросы.

Проведем обработку признаков, чтобы избавиться от шума, который может повлиять на результаты работы алгоритмов МО. При этом, по возможности, сохраним обработанные данные в новую переменную, чтобы проверить влияние очистки на итоговые оценки модели.

```python
# подготовим датасет
df_cleaned = df.copy()
```

Посмотрим на выбросы:

```python
df_count_out = df.copy()
```

```python
for x in df_count_out.columns:
    q75, q25 = np.percentile(df_count_out.loc[:, x], [75, 25])
    qr = q75 - q25
    max = q75 + 1.5 * qr
    min = q25 - 1.5 * qr
    df_count_out.loc[df_count_out[x] < min, x] = np.nan
    df_count_out.loc[df_count_out[x] > max, x] = np.nan
df_count_out.isnull().sum()
```

```python
#Создадим функцию для рисования боксплотов
def boxplot_slave(data, name, xlim = None, ylim = None):
    '''
    boxplot_slave принимает данные в параметр data и возвращает боксплот;
    name используется для разметки и принимает название исследуемого признака.
    xlim и ylim используются для масштабирования графика, ограничивая его оси
    и принимают кортеж (tuple) из двух цифр (начало и конец ограничения)
    '''
    # выведем описание данных
    print('\033[1m' + f'Описание признака {name}:' + '\033[0m')
    print(data.describe())
    print('=' * 45)
    
    # построим боксплот
    box = plt.boxplot(
        x=data, vert=False, 
        patch_artist=True, notch=True
    )
    fig = plt.gcf()
    fig.set_size_inches(15, 8)
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    plt.title(
            f'Боксплот признака {name}', 
            fontsize=14
        )
    plt.show()
    
    # выведем его минимальные и максимальные значения
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - (1.5 * IQR)
    upper_fence = Q3 + (1.5 * IQR)
    print('\033[1m' + f'Пороговые значения боксплота {name}:' + '\033[0m')
    print('Нижняя граница значений:', lower_fence)
    print('Верхняя граница значений:', upper_fence)
    print()
    print()
```
