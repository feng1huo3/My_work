Сосредоточьте цель, чтобы избежать начального смещения повышения градиента: повышение градиента с помощью метода «рекурсии» не учитывает начальную оценку (здесь средняя цель по умолчанию).

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('data_model.csv', index_col = 0)
X = df.drop('T1', axis = 1)
y = df['T1']

y -= y.mean()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

Давайте подгоним a MLPRegressor и вычислим графики частичной зависимости с одной переменной.

```python
from time import time
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import QuantileTransformer
from sklearn.neural_network import MLPRegressor

print("Training MLPRegressor...")
tic = time()
est = make_pipeline(
    QuantileTransformer(),
    MLPRegressor(
        hidden_layer_sizes=(30, 15),
        learning_rate_init=0.01,
        early_stopping=True,
        random_state=0,
    ),
)
est.fit(X_train, y_train)
print(f"done in {time() - tic:.3f}s")
print(f"Test R2 score: {est.score(X_test, y_test):.2f}")
```

Мы настроили конвейер для масштабирования входных числовых функций и настроили размер нейронной сети и скорость обучения, чтобы получить разумный компромисс между временем обучения и прогнозируемой производительностью на тестовом наборе.

Важно отметить, что этот набор табличных данных имеет очень разные динамические диапазоны для своих функций. Нейронные сети, как правило, очень чувствительны к функциям с различными масштабами, и если вы забудете предварительно обработать числовую функцию, это приведет к очень плохой модели.

Можно было бы получить еще более высокую прогностическую эффективность с более крупной нейронной сетью, но обучение также было бы значительно дороже.

Обратите внимание, что важно проверить, достаточно ли точна модель на тестовом наборе, прежде чем строить частичную зависимость, поскольку объяснение влияния данной функции на функцию предсказания плохой модели будет малополезным.

Построим частичную зависимость, как индивидуальную (ICE), так и усредненную (PDP). Мы ограничиваемся только 50 кривыми ICE, чтобы не перегружать график.

```python
import matplotlib.pyplot as plt
from sklearn.inspection import partial_dependence
from sklearn.inspection import PartialDependenceDisplay

print("Computing partial dependence plots...")
tic = time()
features = ["F2", "F7", "F8", "F9"]
display = PartialDependenceDisplay.from_estimator(
    est,
    X_train,
    features,
    kind="both",
    subsample=50,
    n_jobs=3,
    grid_resolution=20,
    random_state=0,
    ice_lines_kw={"color": "tab:blue", "alpha": 0.2, "linewidth": 0.5},
    pd_line_kw={"color": "tab:orange", "linestyle": "--"},
)
print(f"done in {time() - tic:.3f}s")
display.figure_.suptitle(
    "Partial dependence of T1 on the features\n"
    "for the dataset, with MLPRegressor"
)
display.figure_.subplots_adjust(hspace=0.3)
```

Давайте теперь подгоним a HistGradientBoostingRegressor и вычислим частичную зависимость от тех же признаков.

```python
from sklearn.ensemble import HistGradientBoostingRegressor

print("Training HistGradientBoostingRegressor...")
tic = time()
est = HistGradientBoostingRegressor(random_state=0)
est.fit(X_train, y_train)
print(f"done in {time() - tic:.3f}s")
print(f"Test R2 score: {est.score(X_test, y_test):.2f}")
```

Здесь мы использовали гиперпараметры по умолчанию для модели повышения градиента без какой-либо предварительной обработки, поскольку древовидные модели естественным образом устойчивы к монотонным преобразованиям числовых признаков.

Обратите внимание, что в этом табличном наборе данных Gradient Boosting Machines значительно быстрее обучаются и более точны, чем нейронные сети. Также значительно дешевле настроить их гиперпараметры (значения по умолчанию, как правило, работают хорошо, в то время как для нейронных сетей это нечасто).

Построим частичную зависимость, как индивидуальную (ICE), так и усредненную (PDP). Мы ограничиваемся только 50 кривыми ICE, чтобы не перегружать график.

```python
print("Computing partial dependence plots...")
tic = time()
display = PartialDependenceDisplay.from_estimator(
    est,
    X_train,
    features,
    kind="both",
    subsample=50,
    n_jobs=3,
    grid_resolution=20,
    random_state=0,
    ice_lines_kw={"color": "tab:blue", "alpha": 0.2, "linewidth": 0.5},
    pd_line_kw={"color": "tab:orange", "linestyle": "--"},
)
print(f"done in {time() - tic:.3f}s")
display.figure_.suptitle(
    "Partial dependence of T1 on the features\n"
    "for the dataset, with MLPRegressor"
)
display.figure_.subplots_adjust(wspace=0.4, hspace=0.3)
```

Мы можем ясно видеть на PDP (пунктирная оранжевая линия), что медианная цена дома показывает линейную зависимость от среднего дохода (вверху слева) и что цена дома падает, когда увеличивается среднее количество жильцов на домохозяйство (вверху посередине). Верхний правый график показывает, что возраст дома в районе не оказывает сильного влияния на (медианную) цену дома; то же самое и со средним количеством комнат на домохозяйство.

Кривые ICE (светло-голубые линии) дополняют анализ: мы видим, что есть некоторые исключения, когда цена дома остается постоянной при среднем доходе и среднем количестве жильцов. С другой стороны, хотя возраст дома (вверху справа) не оказывает сильного влияния на среднюю среднюю цену дома, похоже, есть ряд исключений, когда цена дома увеличивается в возрасте от 15 до 25 лет. Аналогичные исключения можно наблюдать и для среднего количества комнат (внизу слева). Таким образом, графики ICE показывают некоторый индивидуальный эффект, который ослабляется при использовании средних значений.

На всех графиках галочки на оси x представляют децили значений признаков в обучающих данных.

Мы также наблюдаем, что MLPRegressorимеет гораздо более гладкие прогнозы, чем HistGradientBoostingRegressor.

Важно!!! мы создаем потенциально бессмысленные синтетические образцы, если признаки коррелируют.

2D
PDP с двумя интересующими функциями позволяют нам визуализировать взаимодействие между ними. Однако ICE не может быть простым способом нанесен на график и, таким образом, интерпретирован. Другое соображение связано с производительностью для вычисления PDP. С древовидным алгоритмом, когда запрашиваются только PDP, они могут быть вычислены эффективным способом с использованием 'recursion'метода.

```python
features = ["F2", "F7", "F8", "F9"]
print("Computing partial dependence plots...")
tic = time()
_, ax = plt.subplots(ncols=4, figsize=(9, 4))
display = PartialDependenceDisplay.from_estimator(
    est,
    X_train,
    features,
    kind="average",
    n_jobs=2,
    grid_resolution=10,
    ax=ax,
)
print(f"done in {time() - tic:.3f}s")
display.figure_.suptitle(
    "Partial dependence of T1 on the features\n"
    "for the dataset, with Gradient Boosting"
)
display.figure_.subplots_adjust(wspace=0.4, hspace=0.3)
```

Двухсторонний график частичной зависимости показывает зависимость медианной цены дома от совокупных значений возраста дома и среднего количества жильцов на домохозяйство. Мы можем ясно видеть взаимодействие между двумя характеристиками: при средней заполняемости более двух цена дома почти не зависит от возраста дома, тогда как при значениях менее двух наблюдается сильная зависимость от возраста.

Давайте сделаем такой же график частичной зависимости для взаимодействия двух признаков, но на этот раз в трех измерениях.

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

features = ("F2", "F7")
pdp = partial_dependence(
    est, X_train, features=features, kind="average", grid_resolution=10
)
XX, YY = np.meshgrid(pdp["values"][0], pdp["values"][1])
Z = pdp.average[0].T
ax = Axes3D(fig)
fig.add_axes(ax)

surf = ax.plot_surface(XX, YY, Z, rstride=1, cstride=1, cmap=plt.cm.BuPu, edgecolor="k")
ax.set_xlabel(features[0])
ax.set_ylabel(features[1])
ax.set_zlabel("Partial dependence")
# pretty init view
ax.view_init(elev=22, azim=122)
plt.colorbar(surf)
plt.suptitle(
    "Partial dependence of T1 on median\n"
    "age and average occupancy, with Gradient Boosting"
)
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
X, y = make_regression(n_samples=200, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
# Don't cheat - fit only on training data
scaler.fit(X_train)  
X_train = scaler.transform(X_train)  
# apply same transformation to test data
X_test = scaler.transform(X_test)  
```

```python
regr = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
regr.predict(X_test[:2])
regr.score(X_test, y_test)
```

Многослойный Perceptron чувствителен к масштабированию функций, поэтому настоятельно рекомендуется масштабировать ваши данные. Например, масштабируйте каждый атрибут входного вектора X до [0, 1] или [-1, +1] или стандартизируйте его, чтобы он имел среднее значение 0 и дисперсию 1. Обратите внимание, что вы должны применить такое же масштабирование к тестовому набору для значимые результаты. Вы можете использовать StandardScalerдля стандартизации.
