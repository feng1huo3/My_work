```python
import numpy as np
import pandas as pd

from keras.models import Sequential
from keras.layers import Dense

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('data_model.csv', index_col = 0)
df.sample(5)
```

```python
df_1 = df.drop(['F1', 'F4', 'F5', 'F6', 'T2', 'F7', 'F8', 'F10'], axis = 1)
```

```python
ds = df.values
X = ds[:,1:4]
y = ds[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state = 42)

#res= clf._predict_proba_lr(X_test,Y_test)

X_train.shape
```

```python
# Среднее значение
mean = X_train.mean(axis=0)
# Стандартное отклонение
std = X_train.std(axis=0)
X_train -= mean
X_train /= std
X_test -= mean
X_test /= std
```

Создали сеть

```python
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(X_train.shape[1],)))
#model.add(Dense(16, activation='relu', input_shape=(X_train.shape[1],)))
#model.add(Dense(3, activation='relu', input_shape=(X_train.shape[1],)))
#model.add(Dense(3, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

Обучаем

```python
model.fit(X_train, y_train, epochs=20, batch_size=1, verbose=2)
```

Оценка точности работы сети

```python
mse, mae = model.evaluate(X_test, y_test, verbose=0)
```

```python
print("Средняя абсолютная ошибка (ед.):", mae)
```

Использование сети для предсказания соотношения

```python
pred = model.predict(X_test)
```

```python
print("Предсказанное соотношение:", pred[1][0], ", правильное соотношение:", y_test[1])
```

```python
# Предсказание vs правильный ответ
for i in range(len(pred)):
  print("Сеть сказала: ", np.round(pred[i],2), ", а верный ответ: ", np.round(y_test[i],2), ", разница: ", np.round(pred[i] - y_test[i],2))
```
